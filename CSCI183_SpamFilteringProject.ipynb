Write a function to get rid of commas and periods in text

```python
def remove_punctuation(df):
    df = df.copy()
    for col in df:
        df[col] = df[col].str.replace(',','')
        df[col] = df[col].str.replace('.','')
        df[col] = df[col].str.replace('!','')
        df[col] = df[col].str.replace(':','')
        df[col] = df[col].str.replace('?','')
    return df
```
